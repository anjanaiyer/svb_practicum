```python
df = pd.read_csv('buyouts_lp_csv/buyouts_lp_orig.csv')
LP_ID = df['ID'].to_list()
```

```python
df1 = pd.DataFrame(columns =['lp_id'])
```

```python
for i in LP_ID:
    first_part = ('https://ra.pei.blaize.io/api/v2/buyouts/institutions/')
    url = (first_part + str(i) + "/contacts")
    r = requests.get(url, timeout=(10,200), headers=headers, cookies=cookies)
    s = bs(r.content, 'lxml')
    txt = s.get_text()
    if not txt:
        print("no data in " + str(i))
        continue
    df1 = df1.append({'lp_id':i}, ignore_index=True)
    dict2 = json.loads(txt)
    df1 = df1.append(pd.json_normalize(dict2['data']),ignore_index=True)
    print(str(i) + " done")
```

```python
df1.head()
```

```python
df1.lp_id.fillna(method='ffill', inplace = True)
```

```python
 df1['lp_id'] = df1['lp_id'].apply(np.int64) 
```
