```python
df = pd.read_csv('buyouts_funds_csv/buyouts_funds_orig.csv')
```

```python
fm_id = df['FundManagerID'].to_list()
```

```python
df1 = pd.DataFrame()
```

```python
fm_id_list =[]
```

```python
for i in fm_id:
    first_part = ('https://ra.pei.blaize.io/api/v2/buyouts/institutions/')
    url = (first_part + str(i) + "/appetite")
    r = requests.get(url, timeout=(10,200), headers=headers, cookies=cookies)
    s = bs(r.content, 'lxml')
    txt = s.get_text()
    if not txt:
        print("no data in " + str(i))
        continue
    fm_id_list.append(i)    
    dict2 = json.loads(txt)
    df1 = df1.append(pd.json_normalize(dict2['data']),ignore_index=True)
    print(str(i) + " done")
```

```python
df1.head()
```

```python
df1['fm_id'] = fm_id_list
```

```python
df1_strategy= df1.filter(['fm_id','fundStrategy'],axis = 1)
```

```python
df1_sector = df1.filter(['fm_id','sectors'], axis = 1)
```

```python
df1_strategy = df1_strategy.explode('fundStrategy')
```

```python
df1_sector = df1_sector.explode('sectors')
```

```python
df1_strategy = df1_strategy.join(df1_strategy['fundStrategy'].apply(pd.Series), rsuffix="_l2").explode('regions')
```

```python
df1_sector = df1_sector.join(df1_sector['sectors'].apply(pd.Series), rsuffix="_l3")
```

```python
df1_sector.head()
```

```python
df1_sector.drop(columns=['sectors',0], axis = 1, inplace = True)
```

```python
list_col = df1_sector.columns.tolist()
```

```python
list_col.insert(df1_sector.columns.get_loc('hasAppetite') ,list_col.pop(-1))
```

```python
df1_sector = df1_sector.reindex(columns= list_col)
```

```python
df1_strategy.drop(columns=['fundStrategy',0], axis = 1, inplace = True)
```

```python
df1_sector.to_csv('buyouts_funds_csv/fm_sectors.csv', index = False)
```

```python
df1_strategy.to_csv('buyouts_funds_csv/fm_strategy.csv', index = False)
```
